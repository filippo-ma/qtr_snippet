import pandas as pd
import numpy as np

```python
#load data
nyse_banks = pd.read_excel('nyse_banks_origin.xlsx')
```

```python
#pd.set_option('display.max_row', len(nyse_banks.columns))
nyse_banks.transpose()
```

```python
q2_2020 = '2020 Quarter 2'
q1_2020 = '2020 Quarter 1'
q4_2019 = '2019 Quarter 4'
q3_2019 = '2019 Quarter 3'
q2_2019 = '2019 Quarter 2'
q1_2019 = '2019 Quarter 1'
q4_2018 = '2018 Quarter 4'
q3_2018 = '2018 Quarter 3'
q2_2018 = '2018 Quarter 2'
q1_2018 = '2018 Quarter 1'
q4_2017 = '2017 Quarter 4'

qtr_array = np.array([q2_2020, q1_2020,
                      q4_2019, q3_2019, q2_2019, q1_2019,
                      q4_2018, q3_2018, q2_2018, q1_2018, 
                      q4_2017]) 

date_2020_q2 = '2020-06-30'
date_2020_q1 = '2020-03-31'
date_2019_q4 = '2019-12-31'
date_2019_q3 = '2019-09-30'
date_2019_q2 = '2019-06-30'
date_2019_q1 = '2019-03-31'
date_2018_q4 = '2018-12-31'
date_2018_q3 = '2018-09-30'
date_2018_q2 = '2018-06-30'
date_2018_q1 = '2018-03-31'
date_2017_q4 = '2017-12-31'

# add qtr end date 
nyse_banks['date 2020 Quarter 2'] = date_2020_q2
nyse_banks['date 2020 Quarter 1'] = date_2020_q1
nyse_banks['date 2019 Quarter 4'] = date_2019_q4
nyse_banks['date 2019 Quarter 3'] = date_2019_q3
nyse_banks['date 2019 Quarter 2'] = date_2019_q2
nyse_banks['date 2019 Quarter 1'] = date_2019_q1
nyse_banks['date 2018 Quarter 4'] = date_2018_q4
nyse_banks['date 2018 Quarter 3'] = date_2018_q3
nyse_banks['date 2018 Quarter 2'] = date_2018_q2
nyse_banks['date 2018 Quarter 1'] = date_2018_q1
nyse_banks['date 2017 Quarter 4'] = date_2017_q4


df = []
for qtr in qtr_array:    
    column_names = ['ticker','date', 'share type', 'noninterest income', 'interest income', 'interest expense',
                    'net interest income', 'provision for credit losses','noninterest expense', 
                    'income before tax', 'tax expense', 'net income']

    df_temp = nyse_banks[['Ticker symbol',
                          'date {}'.format(qtr),
                          'Tipo di azione', 
                          'Altri proventi di gestione\nUSD {}'.format(qtr), 
                          'Interessi attivi\nUSD {}'.format(qtr), 
                          'Interessi passivi\nUSD {}'.format(qtr), 
                          'Interessi netti\nUSD {}'.format(qtr),  
                          'Fondi perdite su crediti\nUSD {}'.format(qtr),
                          'Spese generali\nUSD {}'.format(qtr),
                          'Utile lordo\nUSD {}'.format(qtr), 
                          'Imposte\nUSD {}'.format(qtr), 
                          'Utile netto\nUSD {}'.format(qtr),
                         ]]
    
    df_temp.columns=column_names
    df.append(df_temp)

```

```python
dfc = pd.concat(df)
```

```python
ticks = dfc.ticker.unique()
```

```python
dfs = {}
for tick in ticks:
    df_temp = pd.DataFrame(dfc.groupby('ticker').get_group(tick))
    dfs[tick] = df_temp
```

```python
banks_income_statement_qtr = pd.concat(dfs)
```

```python
banks_income_statement_qtr.replace('n.d.', np.nan, inplace=True)
```
